```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
X=2-3*np.random.normal(0,1,20)
y=X-2*(X**2)+0.5*(X**3)+np.random.normal(-3,3,20)
plt.scatter(X,y,s=10)
plt.show()
```

```python
X.shape
```

```python
X=X.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
y_pred=model.predict(X)
from sklearn.metrics import r2_score,root_mean_squared_error
r2=r2_score(y,y_pred)
rmse=root_mean_squared_error(y,y_pred)
print(r2)
print(rmse)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred,color="Red")
```

```python
from sklearn.preprocessing import PolynomialFeatures
degree=4

polyfit=PolynomialFeatures(degree=degree)
```

```python
X_poly=polyfit.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression
mod=LinearRegression()
mod.fit(X_poly,y)
```

```python
y_pred=mod.predict(X_poly)
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
print(r2_score(y_pred,y))
print(root_mean_squared_error(y_pred,y))
```

```python
sorted_idx = X[:,0].argsort()
plt.scatter(X,y)
plt.plot(X[sorted_idx], y_pred[sorted_idx], color='red')


```
