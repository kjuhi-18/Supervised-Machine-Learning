```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
x=np.random.randint(1,50,20)
```

```python
x
```

```python
x = np.sort(x)
print(x)
```

```python
y = 3*(x**2) + 2*x + 5 + np.random.randint(-50,50,size=len(x))
```

```python
x=x.reshape(-1,1)
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
y_pred=model.predict(x)
from sklearn.metrics import r2_score,root_mean_squared_error
r2=r2_score(y,y_pred)
rmse=root_mean_squared_error(y,y_pred)
print(r2)
print(rmse)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color="Red")
```

```python
from sklearn.preprocessing import PolynomialFeatures
degree=2
polyfit=PolynomialFeatures(degree=degree)
```

```python
X_poly=polyfit.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
mod=LinearRegression()
mod.fit(X_poly,y)
```

```python
y_pred=mod.predict(X_poly)
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
print(r2_score(y_pred,y))
print(root_mean_squared_error(y_pred,y))
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color="Red",linewidth=2)
```
