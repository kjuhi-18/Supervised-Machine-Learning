```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('regression.csv')
```

```python
# dimension of df
print(df.shape)
```

```python
# Declare feature variable and target variable
X = df['SAT'].values # Sales and Advertising data values are given by X
y = df['GPA'].values # Sales and Advertising data values are given by Y
```

```python
#print the dimensions of x and y
X.shape
```

```python
y.shape
```

```python
#reshaping x and y
X=X.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
#Split X and y into training and test data sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=42)
X_VAL,X_test1,y_VAL,y_test1 = train_test_split(X_test,y_test,test_size=0.5, random_state=42)
```

```python
# print the dimensions of X_train, X_test, y_train, y_test
print(X_train.shape)
print(X_VAL.shape)
print(X_test1.shape)
print(y_train.shape)
print(y_VAL.shape)
print(y_test1.shape)
```

```python
#fit the linear model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
y_pred1=model.predict(X_VAL)
```

```python
#compute model slope and intercept
a=model.coef_
b=model.intercept_
print("Estimated Model Coefficient:",a)


print("Estimated intercept of Model:",b)
```

```python
#calculate the root mean squared error
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_VAL,y_pred1)
rmse=np.sqrt(mse)
print("RMSE:",rmse)
```

```python
#caluclating the r2 score
from sklearn.metrics import r2_score
print(r2_score(y_VAL,y_pred1))
```

```python
#plot the regression line
plt.scatter(X,y,color="blue",label="scatter plot")
plt.plot(X_VAL,y_pred1,color="black",label="regression line")
plt.title('Relationship between SAT and GPA')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.legend(loc=4)
plt.show()
```

```python
y_pred=model.predict(X_test1)
```

```python
#caluclating the r2 score
from sklearn.metrics import r2_score
print(r2_score(y_test1,y_pred))
```

```python
#plot the regression line
plt.scatter(X,y,color="blue",label="scatter plot")
plt.plot(X_test1,y_pred,color="black",label="regression line")
plt.title('Relationship between Sales and Advertising')
plt.xlabel('Sales')
plt.ylabel('Advertising')
plt.legend(loc=4)
plt.show()
```
