```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Salary_Data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X=df['YearsExperience'].values
y=df['Salary'].values
```

```python
plt.scatter(X,y,color='red',label='Scatter Plot')
plt.title("Relationship between Experience and Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend(loc=4)
plt.show()
```

```python
print(X.shape)
print(y.shape)
```

```python
X=X.reshape(-1,1)
y=y.reshape(-1,1)
#Printing the dimensions after reshaping
print(X.shape)
print(y.shape)
```

```python
#Split X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=42)
```

```python
#print the dimensions of X_train, X_test, y_train, y_test
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Fit linear model
#Instantiate the linear regression object lm
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

#train the model using training data sets
lm.fit(X_train,y_train)

#predict on the test data
y_pred = lm.predict(X_test)
```

```python
#Compute model slope and intercept

a= lm.coef_
b=lm.intercept_
print("Estimated model slope, a:", a)
print("Estimated model intercept, b:", b)
```

```python
lm.predict(X)
```

```python
#Calculate and print root mean square error (RMSE)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("Root Mean Square Error: {:.4f}".format(rmse))
```

```python
#r2 score
from sklearn.metrics import r2_score
print("r2 score: {:.4f}".format(r2_score(y_test, y_pred)))
```

```python
#plot of refression line

plt.scatter(X,y,color='blue',label = 'Scatter Plot')
plt.plot(X_test,y_pred,label='Regression Line')
plt.title('Relationship between Experience and Salary')
plt.xlabel('Years Experience')
plt.ylabel('Salary')
plt.legend(loc=4)
plt.show()
```

```python
#check overfitting or underfitting of the data
print("Training set score : {:.4f}".format(lm.score(X_train, y_train)))
print("Test set score : {:.4f}".format(lm.score(X_test, y_test)))
```
